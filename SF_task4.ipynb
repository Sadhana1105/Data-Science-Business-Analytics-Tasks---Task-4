```python
mmcol = (round(gtdata.isnull().sum()/len(gtdata)*100 , 2)>30)
```

```python
mmcol
```

```python
#To drop the columns which have more than 30% of missing values

missingcols = pd.DataFrame((mmcol))
missingcols=missingcols[missingcols[0]==True].index
gtdata.drop(columns=missingcols,axis=1,inplace=True)
```

```python
gtdata.drop(['eventid','extended','latitude','longitude','specificity','vicinity','crit2','crit3','doubtterr','multiple','corp1'
             ,'gname','guncertain1','individual','weapsubtype1','INT_LOG', 'INT_IDEO', 'INT_MISC', 'INT_ANY'],axis=1,
            inplace=True)
```

```python
gtdata.columns
```

```python
gtdata.rename(columns={'iyear':'Year', 'imonth':'Month', 'iday':'Date','country':'CountryCode', 'country_txt':'Country',
                       'region':'regioncode', 'region_txt':'Region','provstate':'State','city':'City', 'crit1':'Critical', 
                       'success':'Success', 'suicide':'Suicide','attacktype1':'attacktypecode', 'attacktype1_txt':'AttackType',
                       'targtype1':'targetypecode', 'targtype1_txt':'TargetType','targsubtype1_txt':'TargetSubType',
                       'targsubtype1':'targetsubcode', 'target1':'Target','natlty1':'nationalitycode','natlty1_txt':'Nationality'
                       ,'weaptype1':'Weaponcode','weaptype1_txt':'WeaponType','weapsubtype1_txt':'WeaponSubType','nkill':'Killed',
                       'nwound':'wounded'},inplace=True)
```

```python
gtdata.head(15)
```

```python
gtdata.isnull().sum()
```

```python
#Replacing the Missing values for Numerical data 

gtdata.Killed.fillna(gtdata.Killed.mode()[0],inplace=True)
gtdata.wounded.fillna(gtdata.wounded.mode()[0],inplace=True)
```

```python
gtdata.ishostkid.fillna(gtdata.ishostkid.mode()[0],inplace=True)
```

```python
gtdata.fillna('Not Specified',inplace=True)
```
